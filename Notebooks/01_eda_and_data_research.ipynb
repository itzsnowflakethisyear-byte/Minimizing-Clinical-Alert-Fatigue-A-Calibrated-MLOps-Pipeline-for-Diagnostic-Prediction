```python
# Correcting the 'Name' column to have proper title case
patient_df['Name'] = patient_df['Name'].str.title()
```

```python
# Converting 'Admission_Date' and 'Discharge_Date' to datetime format
patient_df['Date of Admission'] = pd.to_datetime(patient_df['Date of Admission'] )
patient_df['Discharge Date'] = pd.to_datetime(patient_df['Discharge Date'])
```

```python
# Length of Stay calculation
patient_df['Length of Stay'] = (patient_df['Discharge Date'] - patient_df['Date of Admission']).dt.days
```

```python
patient_df['Length of Stay'].describe()
```

```python
# Rounding Billing Amount to 2 decimal places
patient_df['Billing Amount'] = patient_df['Billing Amount'].round(2)
```

```python
# Billing Amount negative value check
(patient_df['Billing Amount'] < 0).sum()
```

```python
# This sorts by Billing Amount and takes the 10 lowest values
negative_billing_df = patient_df.nsmallest(30, 'Billing Amount')
print(negative_billing_df[['Name', 'Billing Amount', 'Medical Condition']])
```
